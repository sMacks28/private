## Controlling programs with "if" and "while"

```python
x = int(input("Enter a whole number no greater than 10."))
if x > 10:                                                     # if condition :
    print("You entered a number greater than 10.")            # tell computer what to do if condition is fulfilled
    print("Let me fix that for you.")                          # --> do all what is indented (any number of spaces)
    x = 5
print("Your number is ",x)                                     # This is out of the if condition
```

More examples:

`if x == 1:`        $\longrightarrow$ Check if $x=1$

`if x > 1:`         $\longrightarrow$ Check if $x>1$

`if x < 1:`        $\longrightarrow$ Check if $x<1$

`if x >= 1:`        $\longrightarrow$ Check if $x\geq1$

`if x <= 1:`        $\longrightarrow$ Check if $x<1$

`if x != 1:`        $\longrightarrow$ Check if $x\neq1$

Combine several conditions:

```python
if x>10 or x< 1:
    print("Your number is either too big or too small.")
```

```python
if x<=10 and x>=1:
    print("Your number is just right.")
```

```python
if x>10:
    print("Your number is greater than 10.")
elif x>9:                                                        # elif = else if
    print("Your number is ok, but you're cutting it close.")
else:
    print("Your number is fine.")
        
```

```python
x = int(input("Enter a whole number no greater than 10."))
while x > 10:
    print("Your number is greater than 10. Please try again.")
    x = int(input("Enter a whole number greater than ten."))
print("Your number is ",x)
```

## Lists

These are containers.

Elements of a **list** can be numbers, strings or both

Example: `[5, 2, -99]`

```python
r = [ 3, 4, 8, 1, 3, 57 ]
print(r)
```

```python
x = 1.0
y = 1.5
z = -2.2
r = [ x, y, z ]
r = [ 2*x, x+y, z/sqrt(x**2+y**2) ]
```

Elements in a list `r` are denoted `r[0]`, `r[1]`, `r[2]`, ..., `r[-2]`, `r[-1]`

```python
from math import sqrt
r = [ 1.0, 1.5, -2.2 ]
length = sqrt( r[0]**2 + r[1]**2 + r[2]**2 )
print(length)
```

```python
r = [ 1.0, 1.5, -2.2 ]
total = sum(r)
print('Total = ', total)
mean = sum(r)/len(r)
print('Mean = ', mean)
```

**Example: Emission lines of hydrogen**

Program to calculate the wavelengths of emission lines in the spectrum of the hydrogen atom, based on the Rydberg formula

$$\frac{1}{\lambda} = R \left(\frac{1}{m^2} - \frac{1}{n^2} \right)$$

```python
R = 1.097e-2
for m in range(1,4):
    print("Series for m =",m)
    for n in range(m+1,m+6):
        invlambda = R*(1/m**2-1/n**2) 
        print("n =",n," : ",1/invlambda," nm")
```

## User-defined functions

Python comes with many many functions, but often you need to define your own function.

Example: factorial of *n*: 
$$n! = \prod_{k=1}^n k$$


```python
def factorial(n):
    f = 1.0
    for k in range(1,n+1):
        f *= k
        print(f)
    return f
```

```python
a = factorial(10)
print('10!=',a)
```

### .h5 files

An HDF5 file is a container for two kinds of objects: datasets, which are array-like collections of data, and groups, which are folder-like containers that hold datasets and other groups. 

Groups work like dictionaries, and datasets work like NumPy arrays

```python
import h5py

f=h5py.File('file.h5','r') #open the file

f_keys=f.keys() #obtain the most external datasets and groups names, say one dataset named key1 and one group named key2

dataset1=f['key1'] #store in the variable dataset1 the content of dataset key1 (which is an array)
group1=f['key2'] #store in the variable group1 the content of group key2 (which is a dictionary)

fg_keys=f['key2'].keys() #obtain the names of groups and datasets in group key2, say only one dataset named key3

dataset2=f['key2']['key3'] #key

f.close() #always close the file after you have done the reading!
```
