```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#R = 1, 2*mu*V0 = 1

def Smatrix(k):

    k0 = np.sqrt(k**2 + 1)
    cot = np.cos(k0)/np.sin(k0)

    numerator = k0*cot + 1j*k
    denominator = k0*cot - 1j*k
    return np.exp(-2*1j*k) * (numerator / denominator)
```

```python
n=500

im_k = np.linspace(-3, 3, n)
k = 1j*im_k

res=Smatrix(k)
mag = np.absolute(res)
brightness=mag/(mag+1)
y_max = 100

plt.figure(figsize=(10, 6))
# Verwende plt.plot, da die Datenstruktur eindimensional ist!
plt.plot(im_k, mag, label='$|S_0(k)|$') 
plt.title('Betrag der S-Matrix entlang der imaginären Achse (Re(k)=0)')
plt.xlabel('Imaginärteil der Wellenzahl $k$ (κ)')
plt.ylabel('$|S_0(k)|$')
plt.ylim(0, y_max)
plt.axvline(x=0, color='gray', linestyle='--') # Markiere die reelle Achse
plt.grid(True, linestyle=':')
plt.show()
```

```python
def root(k):
    k = k[0]
    return 1 + k/(np.sqrt(10-k**2))*np.tan(np.sqrt(10-2*k**2))
```

```python
from scipy.optimize import fsolve

for x0 in [0.1, 0.5, 1.0, 1.2, 1.5]:
    try:
        print(x0, fsolve(root, [x0]))
    except:
        pass
```
