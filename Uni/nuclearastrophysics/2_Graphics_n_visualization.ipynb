Example: graph of the function `sin x`, from `x = 0`  to  `x = 10`

```python
import numpy as np

# x-values are generated with linspace: 100 values from 0 to 10
x = np.linspace(0,10,100)

# sin function is applied
y = np.sin(x)

# figure
plt.plot(x,y)
plt.show()
```

Nicer:

```python

x = np.linspace(0,10,100)
y = np.sin(x)

#plot(x,y)
plt.plot(x,y,"g--")
#plot(x,y,"ko")

# after plot, we can give the potting range (ylim and xlim), axis names:
plt.ylim(-1.1,1.1)
plt.xlabel("x axis")
plt.ylabel("y = sin x")
plt.show()
```

```python
from numpy import *

x  = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
 
# Different ways to especify color (c) and linestyle (ls)    
plt.plot(x, y1, 'b-', label='sin x')
plt.plot(x, y2, color ='g', linestyle='--', label='cos x')

# No line, but markers
plt.plot(x, y3, c='k', ls='None', marker='+', label='tan x')   
    
plt.ylim(-1.1,1.1)
plt.xlabel("x axis")
plt.ylabel("sin x, cos x, tan x")

# Legend with info given in label
plt.legend(loc=0)

plt.show()
```

## Scatter plots

Two dependent variables, e.g. temperature and brightness of stars

One measures temperature and brightness (also called magnitude) for a group of stars and wants to visualize how the two quantities are related: *scatter plot*.

```python
data = np.loadtxt("stars.txt", float)
temp = data[:,0]
mag  = data[:,1]

plt.scatter(temp, mag)

plt.title('The Hertzsprung-Russell diagram')
plt.xlabel("Temperature")
plt.ylabel("Magnitude")
plt.xlim(13000,0)
plt.ylim(20,-5)
plt.show()
```

**Example: Wave interference**


Suppose we drop a pebble in a pond and waves radiate out from the spot where it fell. We could create a simple representation of the physics with a sine wave, spreading out in a uniform circle, to represent the height of the waves at some later time. If the center of the circle is at $x_1$, $y_1$ then the distance $r_1$ to the center from a point $x$, $y$ is

$$r_1 = \sqrt{(x-x_1)^2 +(y-y_1)^2}$$

and the sine wave for the height is

$$\xi_1(x,y) = \xi_0 \sin kr_1$$

where $\xi_0$ is the amplitude of the waves and $k$ is the wavevector, related to the wavelength $\lambda$ by $k = 2\pi/\lambda$.

Now suppose we drop another pebble in the pond, creating another circular set of waves with the same wavelength and amplitude but centered on a different point $x_2$, $y_2$:

$$\xi_2(x,y)= \xi_0 \sin kr_2 \, \,\, \mathrm{with} \, \, \, r_2 = \sqrt{(x−x_2)^2+(y−y_2)^2}.$$

Then, assuming the waves add linearly (which is a reasonable assumption for water waves, provided they are not too big), the total height of the surface at a point $x$, $y$ is

$$\xi(x, y) = \xi_0 \sin kr_1 + \xi_0 \sin kr_2.$$

Suppose the wavelength of the waves is $\lambda = 5$ cm, the amplitude is 1 cm, and the centers of the circles are 20 cm apart. Make an image of the height over a 1 m square region of the pond:

 - create an array of values representing the height $\xi$ at a grid of 500 × 500 points to cover the 1 m square
 (i.e., grid points have a separation of 100/500 = 0.2 cm)
 
 - use that array to make a density plot


```python
wavelength = 5.0
k = 2*np.pi/wavelength
xi0 = 1.0
separation = 20.0      # Separation of centers in cm
side = 100.0           # Side of the square in cm
points = 500           # Number of grid points along each side
spacing = side/points  # Spacing of points in cm


# Calculate the positions of the centers of the circles
x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2

# Make an array to store the heights
xi = np.empty([points,points],float) #alternative: np.zeros([points,points])

# Calculate the values in the array
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = np.sqrt((x-x1)**2+(y-y1)**2)
        r2 = np.sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j] = xi0*np.sin(k*r1) + xi0*np.sin(k*r2)

# Make the plot
plt.imshow(xi,origin="lower",extent=[0,side,0,side])
plt.gray()
plt.show()
```
