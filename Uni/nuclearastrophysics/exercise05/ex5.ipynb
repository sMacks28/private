```python
import sys
import matplotlib.pyplot as plt
from nse_solver.nse import NSESolver
import numpy as np
```

```python
nse_comp = NSESolver()
```

```python
rho = [1e10, 1e11, 1e6, 1e6, 1e6, 1e7, 6.28e9]
temp = [10, 50, 6, 6, 6, 50, 7.26]
Ye = [0.4, 0.3, 0.5, 0.45, 0.65, 0.5, 0.44]

As = []
Zs = []
Xn = []
Xp = []
Xalpha = []
Xlight = []
Xheavy = []
Xi = []
for i in range(0, 7):
    nse_comp.solve(rho[i], temp[i], Ye[i])
    As.append(nse_comp._As)
    Zs.append(nse_comp._Zs)
    Xn.append(nse_comp.getXn())
    Xp.append(nse_comp.getXp())
    Xalpha.append(nse_comp.getXlight())
    Xlight.append(nse_comp.getXalpha())
    Xheavy.append(nse_comp.getXheavy())
    
    Xi = nse_comp.getMassFractions()
    print(sum(Xi)) #convergence check
```

```python
for i in range(0, 7):
    nse_comp.solve(rho[i], temp[i], Ye[i])
    Xi = nse_comp.getMassFractions()
    Ai = nse_comp._As.astype(int) #Check if A is integer
    Zi = nse_comp._Zs

    A_unique_sums = np.bincount(Ai, weights=Xi)
    A_range = np.arange(len(A_unique_sums))
    mask = A_unique_sums > 0 #filter all the zeros (because of the log and bincount)
    
    A_plot = A_range[mask]
    X_A_plot = A_unique_sums[mask]

    Zi_int = Zi.astype(int)                           
    Z_unique_sums = np.bincount(Zi_int, weights=Xi)
    Z_range = np.arange(len(Z_unique_sums))
    z_mask = Z_unique_sums > 0 #filter all the zeros
    
    Z_plot = Z_range[z_mask]
    X_Z_plot = Z_unique_sums[z_mask]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    #Zi vs Xi
    ax1.step(Z_plot, X_Z_plot, where='mid', label=f'Density index {i}')
    ax1.set_title(r'$Z_i$ vs $X(Z)$ (Elemental Abundances)')
    ax1.set_xlabel('$Z_i$')
    ax1.set_ylabel('$X_i$')
    ax1.set_yscale("log")
    ax1.set_ylim(1e-10,1.1)
    ax1.grid(True)

    #Ai vs Xi
    ax2.step(A_plot, X_A_plot, where='mid', label=f'Density index {i}')
    ax2.set_title(r'$A_i$ vs $X(A)$ (Mass Chain Abundances)')
    ax2.set_xlabel('$A_i$')
    ax2.set_ylabel('$X_i$')
    ax2.set_yscale("log")
    ax2.set_ylim(1e-10,1.1)
    ax2.grid(True)

    plt.tight_layout()
    plt.show()
```
