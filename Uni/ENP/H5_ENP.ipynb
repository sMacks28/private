```python
using Plots
using StatsBase
using GLM
using DataFrames

# Daten aus Abbildung 1: J -> E (in MeV)
# 0+ -> 0.0 MeV
# 2+ -> 1.513 MeV (1513 keV)
# 4+ -> 11.35 MeV (11350 keV)
J = [0, 2, 4]
J_factor = J .* (J .+ 1)
E = [0.0, 1.513, 11.35]

# Linearer Fit: E = slope * J(J+1)
df = DataFrame(x = J_factor, y = E)
model = lm(@formula(y ~ 0 + x), df) # Fit durch den Ursprung
slope = coef(model)[1]

# Plot erstellen
scatter(J_factor, E, label="Experimentelle Daten", xlabel="J(J+1)", 
        ylabel="E (MeV)", title="Rotationsbande von 8Be", markersize=6)
plot!(J_factor, predict(model, df), label="Linearer Fit", lw=2)

println("Die Steigung (slope) beträgt: ", slope, " MeV")
println("Moment of Inertia I = ħ^2 / (2 * slope)")
```
